```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('G27 dados csv.csv', sep = ';')
df
```

```python
df_participants = df[['participant_id', 'participant_name', 'age']].drop_duplicates()
df_participants
```

```python
df_studies = df[['study_id', 'title', 'field', 'start_date', 'center_id']].drop_duplicates()
df_studies
```

```python
df_center = df[['center_id', 'center_name', 'location']].drop_duplicates()
df_center
```

```python
df_enrollments = df[['enrollment_date', 'status', 'study_id', 'participant_id', 'consent_status']].drop_duplicates()
df_enrollments
```

```python
print(df_participants['participant_id'].is_unique)
print(df_studies['study_id'].is_unique)
print(df_center['center_id'].is_unique)
```

```python
print(df_participants.dtypes)
print(df_studies.dtypes)
print(df_center.dtypes)
print(df_enrollments.dtypes)
```

```python
df_enrollments
```

```python
df_enrollments['enrollment_id'] = range(1, len(df_enrollments) + 1)
df_enrollments
```

```python
conn = sqlite3.connect('Research.db')
```

```python
df_participants.to_sql('Participants', conn, if_exists='append', index=False)
```

```python
df_studies.to_sql('Studies', conn, if_exists='append', index=False)
```

```python
df_center.to_sql('Center', conn, if_exists='append', index=False)
```

```python
df_enrollments.to_sql('Enrollments', conn, if_exists='append', index=False)
```

```python
conn.close()
```
